```python
def airport_column(airportname,nooftimes) :
    
    airport = []

    for i in range(0,nooftimes):
    
        airport.append(airportname)
        
    return airport
```

```python
def dates_column(filename):
    
    f = open(filename,'r')

    unwanted_columns = ['mintempC', 'maxtempF', 'sunHour', 'mintempF', 'maxtempC','totalSnow_cm','astronomy', 'uvIndex']

    data = json.loads(f.read())

    dataframe = pd.DataFrame(data['data']['weather'])

    dataframes = dataframe.drop(unwanted_columns,axis = 1)

    dataframe = pd.DataFrame(dataframes['hourly'])

    dataframes = dataframes.drop(['hourly'],axis = 1)

    dataframes = dataframes.loc[dataframes.index.repeat(24)].reset_index(drop=True)

    dates = dataframes['date']
    
    return dates
```

```python
def final_dataframe(filename,noofdays,airport):
    
        f = open(filename,'r')

        unwanted_columns = ['mintempC', 'maxtempF', 'sunHour', 'mintempF', 'maxtempC','totalSnow_cm','astronomy', 'uvIndex']

        unwanted_columns1 = ['FeelsLikeF','winddir16Point','FeelsLikeC','DewPointC','windspeedMiles','HeatIndexF','HeatIndexC','weatherIconUrl','WindGustMiles','weatherDesc','tempC','WindChillC']

        data = json.loads(f.read())

        dataframe = pd.DataFrame(data['data']['weather'])

        dataframes = dataframe.drop(unwanted_columns,axis = 1)

        dataframe = pd.DataFrame(dataframes['hourly'])

        dictionary1 = dataframe.iloc[0]['hourly'][0]

        dataframe1 = pd.DataFrame.from_dict(dictionary1)

        dataframess = pd.DataFrame(columns = dataframe1.columns)

        for j in range(0,noofdays):

            dictionary1 = dataframe.iloc[j]['hourly'][0]

            dataframe1 = pd.DataFrame.from_dict(dictionary1)

            dictionary2 = dataframe.iloc[j]['hourly'][1]

            dataframe2 = pd.DataFrame.from_dict(dictionary2)

            dataframe1 = dataframe1.append(dataframe2,ignore_index = True)

            for i in range(2,24) :

                dictionary2 = dataframe.iloc[j]['hourly'][i]

                dataframe2 = pd.DataFrame.from_dict(dictionary2)

                dataframe1 = dataframe1.append(dataframe2,ignore_index = True)

            dataframess = dataframess.append(dataframe1,ignore_index = True)

        dataframess = dataframess.drop(unwanted_columns1,axis = 1).drop_duplicates()
        
        nooftimes = noofdays * 24

        dataframess['Airport'] = airport_column(airport,nooftimes)

        dataframess['Date'] = dates_column(filename)

        return dataframess
```

```python
def json_to_csv(filename,filename1,noofdays,airport) :
    
    f = open(filename,'r')
    
    data = json.loads(f.read())
    
    dataframe = final_dataframe(filename,noofdays,airport)
    
    dataframe.to_csv(filename1)
    
    return dataframe
```
